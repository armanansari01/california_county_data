```python
driver = webdriver.Chrome()
driver.get(website)



name = []
population = []
inc_date = []
govt_form = []
about = []
county_supervisor_list = []

for i in range(1,59):
    x_path = f'/html/body/div[3]/div/div[2]/div/div[1]/div/div[2]/div[3]/div/div/div/div/div[{i}]/div/div/div[2]/span/a/img'
    element = driver.find_element(By.XPATH, x_path)
    element.click()
    county_container = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[2]/div/div[1]/div/div[1]/div/div')
    
    name.append(county_container.find_element(By.XPATH, '//*[@id="node-full"]/div/h1/span[1]').text)
    print("Processing county name : ",name)
    population.append(county_container.find_element(By.XPATH, '//*[@id="node-full"]/div/div[4]/p[1]').text)
    print("Processing county population : ",population)
    inc_date.append(county_container.find_element(By.XPATH, '//*[@id="node-full"]/div/div[4]/p[2]').text)
    print("Processing county date : ",inc_date)
    govt_form.append(county_container.find_element(By.XPATH, '//*[@id="node-full"]/div/div[4]/p[4]').text)
    print("Processing county govt : ",govt_form)
    about.append(county_container.find_element(By.XPATH, '//*[@id="node-full"]/div/div[4]/p[5]').text)
    print("Processing county desc : ",about)

    ul_element = driver.find_element(By.XPATH,'//*[@id="node-full"]/div/div[4]/ul')
    li_elements = ul_element.find_elements(By.XPATH,'li')
    supervisors = [li.text for li in li_elements]
    county_supervisor = ", ".join(supervisors)
    county_supervisor_list.append(county_supervisor)
    print("Processing county supervisors ")

    driver.back()
driver.quit()
```
